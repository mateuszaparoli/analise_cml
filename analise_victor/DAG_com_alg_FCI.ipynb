```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("VIOLBR24_processado.csv")

df.dropna(inplace=True)
df = df.drop(columns=["DEF_MENTAL", "DEF_VISUAL", "DEF_AUDITI", "AG_ENVEN", "AG_FOGO", "AG_CORTE", "AG_OBJETO", "AG_ENFOR", "DEF_FISICA", "TRAN_MENT", "TRAN_COMP"])
df = df.reset_index(drop=True)
df_num = df
df.head()
```

```python
for col in df_num.columns:
    if pd.api.types.is_numeric_dtype(df_num[col]):
        continue

    df_num[col] = df_num[col].astype('category').cat.codes

df_num = df_num.replace([np.inf, -np.inf], np.nan).dropna()
```

```python
df_num.head()
```

```python
df_num.columns
```

```python
data = df_num.to_numpy()

colunas = ['LES_AUTOP', 'CS_SEXO', 'NU_IDADE_N', 'CS_RACA', 'CS_ESCOL_N', 'SG_UF',
       'ID_OCUPA_N', 'SIT_CONJUG', 'ORIENT_SEX', 'IDENT_GEN', 'OUT_VEZES',
       'LOCAL_OCOR', 'HORA_OCOR', 'metodo', 'deficiencia']
```

```python
from causallearn.search.ConstraintBased.FCI import fci

# default parameters
g, edges = fci(data)

# visualization
from causallearn.utils.GraphUtils import GraphUtils

pdy = GraphUtils.to_pydot(g, labels=colunas)
pdy.write_png('DAGFCI.png')
```
